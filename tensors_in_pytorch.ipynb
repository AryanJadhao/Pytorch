# **MATHEMATICAL OPERATION**

**1.** **SCALAR** **OPS**

```python
x = torch.rand(2,2)
x
```

```python
# addition
print("addition : ", x+2)

# subtraction
print("subtraction : ", x-2)

# multiplication
print("multiplication : ", x*2)

# division
print("division : ", x/2)

# int division
print("Int division : ", (x*100) // 2)

# mod
print("mod : ", ((x*100) // 2) % 2)

# power
print("power : ", x**2)
```

**2. ELEMENT WISE OPS**

```python
a = torch.rand(2,3)
b = torch.rand(2,3)

print(a)
print(b)
```

```python
# add
print("addition : ",a + b)

# sub
print("sub : ",a - b)

# mul
print("mul : ",a * b)

# div
print("div : ",a / b)

# mod
print("mod : ",a % b)

# floor div
print("floor div : ",a // b)

# exp
print("exp : ",a ** b)
```

**3. REDUCTION OPS**

```python
e = torch.randint(size=(2,3), low=0, high=10)
e
```

```python
# sum
torch.sum(e)
```

```python
# sum along column
torch.sum(e, dim=0)
```

```python
# sum along rows
torch.sum(e, dim=1)
```

```python
# mean
# torch.mean(e) --> error

torch.mean(e.to(torch.float32))

# do it for rows and columns just give dimensions

# similarly median
```

```python
# similarly used to calculate prod, standard deviation (std), variance (var), argmax, argmin
```

**4.MATRIX OPS**

```python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=1, high=5)

print(f)
print(g)
```

```python
# matrix multiplication
torch.matmul(f,g)
```

```python
# dot product
vec1 = torch.tensor([1,2,3])
vec2 = torch.tensor([4,5,6])
torch.dot(vec1,vec2)
```

```python
# transpose
# 0 -----> 1
torch.transpose(f,0,1)
```

```python
# determinant
p = torch.randint(size=(3,3), low=0,  high=10, dtype=torch.float32)
torch.det(p)
```

**5. COMPARISON OPS**

```python
s = torch.randint(size=(2,3), low=0,  high=10, dtype=torch.float32)
t = torch.randint(size=(2,3), low=0,  high=10, dtype=torch.float32)

print(s)
print(t)
```

```python
# greater than
print(s>t)

# less than
print(s<t)

# equal to
print(s == t)

# less than equal to , greater than equal to
```
