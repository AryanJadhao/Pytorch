```python
import pandas as pd
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
torch.manual_seed(42)
```

```python
df = pd.read_csv('/content/drive/MyDrive/Fashion MNIST dataset/fashion-mnist_train.csv')
df
```

```python
x = df.iloc[:, 1:].values
y = df.iloc[:,0].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
# scaling the features
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
x_train
```

```python
# create customDataset class
class CustomDataset(Dataset):
  def __init__(self, features, labels):
    self.features = torch.tensor(features, dtype=torch.float32)
    self.labels = torch.tensor(labels, dtype=torch.long)

  def __len__(self):
    return len(self.features)

  def __getitem__(self, index):
    return self.features[index], self.labels[index]
```

```python
# create train_dataset object
train_dataset = CustomDataset(x_train, y_train)
```

```python
train_dataset[0]
```

```python
# create test_dataset object
test_dataset = CustomDataset(x_test, y_test)
```

```python
train_dataset[0]
```

```python
# create train and test loader
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
# we are doing these to reduce overfitting
# 1) BatchNormalization
# 2) Dropout
# 3) L2 Regularization

# to improve the accuracy
# 1) 83% -----> when run on cpu
# 2) 89% -----> when run on GPU -------> training acc = 98%
# 3) 89% -----> when reduced overfit -----> training acc = 93%
```

```python
from torch.nn.modules.activation import ReLU
# define NN class
class NeuralNetwork(nn.Module):
  def __init__(self, num_features):

    super().__init__()
    self.model = nn.Sequential(
        nn.Linear(num_features, 128),
        nn.BatchNorm1d(128),
        nn.ReLU(),
        nn.Dropout(p=0.3),

        nn.Linear(128, 64),
        nn.BatchNorm1d(64),
        nn.ReLU(),
        nn.Dropout(p=0.3),

        nn.Linear(64, 10)
    )

  def forward(self, x):
    return self.model(x)
```

```python
learning_rate = 0.01
epochs = 100
```

```python
# instantiate NNclass model
model = NeuralNetwork(x_train.shape[1])
model = model.to(device)

# loss function
criterian = nn.CrossEntropyLoss()

# optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=0.1, weight_decay=1e-4)
```

```python
# training loop

for epoch in range(epochs):

  total_epoch_loss = 0
  for batch_features, batch_label in train_loader:

    # move to GPU
    batch_features = batch_features.to(device)
    batch_label = batch_label.to(device)

    # forward
    out = model(batch_features)

    # loss calculation
    loss = criterian(out, batch_label)

    # back prop
    optimizer.zero_grad()
    loss.backward()

    # update grad
    optimizer.step()

    total_epoch_loss = total_epoch_loss + loss.item()

  avg_loss = total_epoch_loss/len(train_loader)

  print(f"epoch: {epoch+1}, loss: {avg_loss}")
```

```python
# set model into evluation mode
model.eval()
```

```python
# evaluation
total = 0
correct = 0

with torch.no_grad():

  for batch_features, batch_label in test_loader:
    # move to GPU
    batch_features = batch_features.to(device)
    batch_label = batch_label.to(device)

    out = model(batch_features)

    _, pred = torch.max(out.data, 1)

    total = total + batch_label.shape[0]
    correct = correct + (pred == batch_label).sum().item()

print(f" accuracy : {(correct/total)*100}%")
```
